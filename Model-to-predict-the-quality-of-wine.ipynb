# Loading the dataset and data analysis

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
df = pd.read_csv("winequality-red.csv")
```

```python
df.rename(columns = {
    'fixed acidity' : 'Fixed-acidity',
    'volatile acidity': 'Volatile-acidity',
    'citric acid' : 'Citric-acid',
    'residual sugar': 'Residual-sugar',
    'chlorides': 'Chlorides',
    'free sulfur dioxide': 'Free-sulfur-dioxide',
    'total sulfur dioxide': 'Total-sulfur-dioxide',
    'density': 'Density',
    'pH': 'pH',
    'sulphates': 'Sulphates',
    'alcohol': 'Alcohol',
    'quality': 'Quality'
})
```

```python
#plotting the boxplot and the scatter plot for each metrics

fig, ax = plt.subplots(11, 2, figsize=(15, 45))
plt.subplots_adjust(hspace = 0.5)

for i in range(11) :
    # AX 1
    sns.boxplot(x=df.columns[i], data=df, ax=ax[i, 0])
    # Ax 2
    sns.scatterplot(x=df.columns[i], y='quality', data=df, hue='quality', ax=ax[i, 1])
```

```python
df.columns
```

```python
a = df.groupby('quality', as_index=False).mean()
a.columns[0]
```

```python
# plotting the different metrics for quality vs all other columns in the table 
    
for i in a.columns[1:]:
    plt.hist2d(x=a['quality'], y=a[i], cmap=plt.cm.Blues)
    plt.colorbar(label='Counts')
    plt.xlabel('Quality')
    plt.ylabel(i)
    plt.title(f'Histogram of Quality vs {i}')
    plt.show()

   
```

```python
#plotting scatter plot by quality counts

import matplotlib.pyplot as plt
import pandas as pd


count = df.groupby('quality').apply(lambda x: x['sulphates'].count()).reset_index(name='Count')
plt.scatter(x=count['quality'], y=count['Count'])
for i, txt in enumerate(count['Count']):
    plt.annotate(txt, (count['quality'][i], count['Count'][i]), 
                 textcoords="offset points", xytext=(0,10), ha='center')

plt.xlabel('Quality')
plt.ylabel('Count of Quality')
plt.title('Number of quality')
plt.show()
```
